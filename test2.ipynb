```python
import math
import cmath
import random
import itertools
import functools
import statistics
import fractions
import decimal
import sys
import os
from pathlib import Path

import numpy as np
import scipy
import sympy as sym
import matplotlib.pyplot as plt

import dtumathtools
from dtumathtools import dtuplot

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def first_n_primes(n):
    primes = []
    num = 2
    while len(primes) < n:
        if is_prime(num):
            primes.append(num)
        num += 1
    return primes

# Example
primes = first_n_primes(100)
# print(primes)
numbers = []
i_list = []

for i in range(len(primes)):
    a = str(primes[i])[-1]
    numbers.append(int(a))
    i_list.append(i)


one = numbers.count(1)
three = numbers.count(3) - 1
seven = numbers.count(7)
nine = numbers.count(9)

vals = [one, three, seven, nine]
print(vals)
val_x = [1,3,7,9]

plt.bar(val_x, vals)
plt.show()
```
